```python
import conx as cx
```

```python
raw_data = """May 2017: 3 times
June 2017: 5 times
July 2017: 6 times
Oct 2017: 1 time
Dec 2017: 2 times
Jan 2018: 1 time
Feb 2018: 3 times
March 2018: 2 times
April 2018: 9 times
May 2018: 20 times
June 2018: 22 times"""
```

```python
data = [[col.strip() for col in row.split(":")] for row in raw_data.split("\n")]
```

```python
data
```

```python
for row in data:
    row[1] = int(row[1].split()[0])
```

```python
data
```

```python
cx.plot(["", [col[1] for col in data]],
        xs=[col[0] for col in data],
        title="witch hunt tweets",
        xlabel="Month", ylabel="count",
        height=6.5,
        width=9,
        xs_rotation=30)
```
